```python
# Made by RASİN AŞCI
# A user can add tasks to the todo list.
task_detail =  {
        "completed": 'no', 
        "priority": '',
        "due_date": '' }
todo_list = {'task1':task_detail, 'task2': task_detail}
task_name = 'a'

def task_add():
  while True:
    task_name = str(input('Enter task name:'))
    if task_name != '' and task_name not in todo_list:
      todo_list[task_name] = task_detail
      
      is_completed(task_name)
      due_date(task_name)
      priority(task_name)
      return start()
    else:
      print("Please re-enter your task")
        
      
def is_completed(task_name):
  while True:
   
    is_completed = str(input("Is this task completed?: ").lower())
    if is_completed == 'yes':
     
      todo_list[task_name]["completed"] = is_completed
      
      break
    elif is_completed == 'no':
      todo_list[task_name]["completed"] = is_completed
      break
    else:
      print("Please only type yes or no")
  
def due_date(task_name):
  due_date = input("When is the due date: ")
  todo_list[task_name]["due_date"] = due_date
  
def priority(task_name):
  priority = input("Set a priority level 1 to 4: ")
  todo_list[task_name]["priority"] = priority

def update_a_task():
  while True:
    print(todo_list)
    requested_task = input("Which task you want to update")
  
    if requested_task != '' and requested_task in todo_list:
      for key in todo_list.values():
        if requested_task == key:
          todo_list[key]
          print("Task Updated!")
        is_completed(requested_task)
        due_date(requested_task)
        priority(requested_task)
        print('Your task is updated')
        break
      break
    else:
      print("There is no such task please re-enter the name")
   # user can check a task completed or not

def delete_task():
  print(todo_list)
  requested_task = input("Which task you want to delete: ")
  for key in todo_list.keys():
    if requested_task == key:
      del todo_list[key]
      break
      print("Task Deleted!")
  
def start():
  while True:
    print(' What do you want to do?')
    print("""
          * to see all the tasks, type 1
          * to add an new task, type 2
          * to upate an task, type 3
          * to delete an task, type 4
          * to exit, type 0

          """)
    user_choice = str(input('your choice :'))
    if user_choice == '1':
      print(todo_list)
      
    elif user_choice == '2':
      task_add()
      break
    elif user_choice == '0':
      print("""
            Good bye...
            
           Made by Rasin 
            """)
      break
    elif user_choice == '3':
      update_a_task()
    elif user_choice == '4':
      delete_task()
    else:
      print('please select a good choice')

start()
```
